```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
```

```python
def kit(x):
    return np.sin(x)

def kat(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))

axs[0].plot(x, kit(x))
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].set_title('sin(x)')

axs[1].plot(x, kat(x))
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].set_title('cos(x)')

plt.savefig('Prompt2.2.pdf')
plt.show()
```
